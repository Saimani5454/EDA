```python
from pandas import read_csv
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
filename = 'TSNE_data.csv'
dataframe = pd.read_csv(filename)
dataframe
```

```python
array=dataframe.values
X=array[:,1:]
Y=array[:,0]
```

```python
from bioinfokit.visuz import cluster
data_tsne=TSNE(n_components=2,perplexity=50.0).fit_transform(X)
cluster.tsneplot(score=data_tsne)
```

```python
color_class=dataframe['diagnosis'].to_numpy()
cluster.tsneplot(score=data_tsne,colorlist=color_class,legendpos='upper right',legendanchor=(1.15,1))
```

```python
X=dataframe.iloc[:,1:]
Y=dataframe.iloc[:,0:]
```

```python
from bioinfokit.visuz import cluster
data_tsne=TSNE(n_components=2,perplexity=50.0).fit_transform(X)
cluster.tsneplot(score=data_tsne)
```

```python
color_class=dataframe['diagnosis']
cluster.tsneplot(score=data_tsne,colorlist=color_class,legendpos='upper right',legendanchor=(1.15,1))
```

```python
# prompt: can you experiment with diiferent perplexicity and display this viz in the notebook save it

import matplotlib.pyplot as plt
# Define different perplexity values to experiment with
perplexities = [5, 30, 50]

# Iterate through each perplexity value
for perplexity in perplexities:
    print(f"Generating t-SNE plot with perplexity = {perplexity}")
    # Apply t-SNE with the current perplexity
    data_tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42).fit_transform(X)

    # Create a t-SNE plot
    cluster.tsneplot(
        score=data_tsne,
        colorlist=color_class,
        legendpos='upper right',
        legendanchor=(1.15, 1),
        # Removed the 'ax' argument as it's not supported in this bioinfokit version
        # ax=None
    )

    # Add a title to the plot
    plt.title(f't-SNE plot with perplexity = {perplexity}')

    # Save the plot to a file
    plt.savefig(f'tsne_perplexity_{perplexity}.png', bbox_inches='tight')

    # Show the plot
    plt.show()
```
